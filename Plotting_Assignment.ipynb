```python
print(df['co2'].head(5))
co2_ems=df['co2']

''' Using seaborn to plot data from the Co2 column in a histogram '''

sn.set_style('dark')
plt.figure(figsize=(12,6))
sn.histplot(data=co2_ems,color='yellow',edgecolor='orange',bins=20)
plt.xlabel('CO2 Emission')
plt.ylabel('Frequency of Car Emission')
plt.title('Carbon Dioxide Emmision Distribution')
plt.savefig('co2_Distribution.png')
plt.show()
```

***Interpretation***
C02 emmissions between the range of 300 to 500 have the highest frequency in emmissions according to the above histogram. C02 emmissions around 100 have the least frequency in emmissions. CO2 emmission frequencies are at the lowest in the range between 100 to 300 after which they peak and lower again from the range of 500 to 700.

```python
print(df['ghgScore'].head(5))
ghgScore=df['ghgScore']

sn.set_style('dark')
sn.histplot(data=ghgScore,color='green',edgecolor='blue',bins=10)
plt.xlabel('Green House Gas score(ghg)')
plt.ylabel('Frequency')
plt.title('(GHG) Score Distributions in Vehicle Emissions')
plt.savefig('ghgScore_Distribution.png')
plt.show()
```

***Interpretation***
The GHG score is at the highest at the 1000 mark frequency and the lowest at below the 50 mark frequency. The bar heights indicate the frequency in relation to the GHG scores.

```python
print(df['comb'].head(5))
comb=df['comb']

sn.set_style('dark')
sn.histplot(data=comb,color='pink',edgecolor='purple',bins=20)
plt.xlabel('Combined fuel Efficiency')
plt.ylabel('Frequency')
plt.title('Combined Fuel Efficiency Distribution')
plt.savefig('comb_efficiency.png')
plt.show()
```

```python
num_cols = df.select_dtypes(include=['number'])
corr_matrix = num_cols.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sn.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".1f")
plt.title('Correlation Heatmap')
plt.savefig('heatmap.png')
plt.show()
```
